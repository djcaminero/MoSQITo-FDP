<h2 id="First_Point_Header">1. The psychoacoustic hearing model</h2>

Research about the functioning of the auditory system falls within the field of psychoacoustics. <b>Psychoacoustics</b> is the study of sound perception of the auditory system, understood as coding, transmission, processing and reception of the signal. Within this field there are indicators that turn out to be purely empirical, such as the concept related to loudness. <b>Loudness</b> is the sensation of intensity perceived by the ear when excited by a sound pressure wave. It depends on the frequency, sound pressure and duration of the incident sound. The evaluation of loudness is framed in two types of magnitudes, loudness <i>N</i> measured in sone, and loudness level <i>L<sub>N</sub></i>, measured in phons.

In order to study the loudness of a signal a psychoacoustic hearing model has been implemented "comp_loudness.py" according to the one presented in ECMA-418-2. <b>The hearing model</b> consists of a set of sequentially ordered calculations that mimic the sound processing performed by the ear of an incident signal under given conditions, in this case, it describes the loudness excitation in a critical band per Bark. The output of the signal consists in a pair of array lists, one for specific loudness and the other one for total loudness. Furthermore, in the function exists an argument that allows a validation of the intermediate steps of the model. For the final model validation, the use of the function "hearing_model_validation" inside "hearing_model_validation.py" is needed.

<h3 id="Third_Sub_Point_1">3.1 Loudness of a pure tone</h3>

First, the modules that are going to be used are imported into the code. Second, we create the pure tone that is analyzed and of which loudness is calculated. Third, the calculation of loudness takes place. The last parameter of the function,"validation", decides whether validation is done or not. If "validation" is set to <i>TRUE</i>, it displays the result of each section that appears on ECMA-418-2 section 5.

```python
# FIRST SECTION - PACKAGE IMPORTS
from mosqito.functions.hearing_model.sine_wave_generator import sine_wave_generator
from mosqito.functions.hearing_model.comp_loudness import comp_loudness

# SECOND SECTION - SINE WAVE CREATION
fs = 48000
p_ref = 2e-5
duration = 1.00
db_SPL = 40.00
sin_freq = 1000.00

sig, time = sine_wave_generator(fs, duration, db_SPL, sin_freq)

# THIRD SECTION - LOUDNESS CALCULATION
s_l, t_l = comp_loudness(sig, validation=True)
```

<b>The output</b> contains all the parameters that are necessary in order to follow into further sections, such us the AFB coefficients and loudness values.

With the previous loudness values that are uncorrected, we take them and we reassign them based on  the input/output relationship of the loudness function for a series of pure tones that  ranges their inputs from 0 dB SPL to 100 dB SPL. This is done as a temporal solution that fixes the compression that is seen in the loudness values.

```python
mean_uncorrected = float(np.mean(t_l))
print('\nMean value, total loudness (Uncorrected): ' + str(mean_uncorrected))
print('Mean value in phons, total loudness (Uncorrected): ' + str(sone2phone(mean_uncorrected)) + '\n\n')

print('\nLOUDNESS CORRECTION')
mean_corrected = reassignment_loudness(mean_uncorrected, mean_loudness_values())
print('\nMean value, total loudness (Corrected): ' + str(mean_corrected))
print('Mean value in phons, total loudness (Corrected): ' + str(sone2phone(mean_corrected)))
```

<b>The output loudness (in phones) of a pure tone at 1 kHz must have the same value than its input sound pressure level</b>. In this case, the sound pressure level of the tone is 40 dB SPL, which corresponds to its loudness, which is 39.962 phons.

<h3 id="Third_Sub_Sub_Point_1">3.1.1 Check the frequency response of a pure tone</h3>


In case we are not sure of the sound pressure level that corresponds to the pure tone we want to be analyzed by the loudness calculation function, we can use the next section to check its value.

```python
import numpy as np
import matplotlib.pyplot as plt
from mosqito.functions.hearing_model.sine_wave_generator import sine_wave_generator


fs = 48000
p_ref = 2e-5
duration = 1.00
db_SPL = 40.00
sin_freq = 1000.00

sig, time = sine_wave_generator(fs, duration, db_SPL, sin_freq)

n = len(sig)
pressure_rms = p_ref * (10.00 ** (db_SPL / 20.00))
sensitivity = np.sqrt(2) * pressure_rms

# Calculations for the level from time domain signal
#
rms_time = np.sqrt(np.mean(np.power(sig, 2)))
db_time = 20 * np.log10(abs(np.fft.fft(sig * np.blackman(n))))
window = np.blackman(n)
signal = sig * window

spectrum = np.fft.fftshift(np.fft.fft(signal))
freq = np.fft.fftshift(np.fft.fftfreq(n, 1 / fs))

spectrum = spectrum[n // 2:]
freq = freq[n // 2:]

spectrum *= 2

if n % 2 == 0:
    spectrum[-1] /= 2

spectrum_mag = np.abs(spectrum) / np.sum(window)

# To obtain RMS values, divide by sqrt(2)
spectrum_rms = spectrum_mag / np.sqrt(2)
# Do not scale the DC component
spectrum_rms[0] *= np.sqrt(2) / 2

# Convert to decibel scale
spectrum_db = 20 * np.log10(spectrum_rms / p_ref)

plt.figure(figsize=(10, 5))
plt.semilogx(freq, spectrum_db)
plt.xlim((1, fs / 2))
plt.ylim((0, 120))
plt.grid('on')
plt.xlabel('Frequency [Hz]')
plt.ylabel('SPL [dB]')
plt.title('Input Frequency Response')
plt.grid(which='both', linestyle='-', color='grey')
plt.show()
```
